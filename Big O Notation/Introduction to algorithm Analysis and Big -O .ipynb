```python
def sum1(n):
    final_sum=0
    
    for x in range(n+1):
        final_sum+=x
    return final_sum
```

```python
sum1(10)
```

```python
def sum2(n):
    return(n*(n+1))/2
```

```python
sum2(10)
```

```python
%timeit sum1(100)
```

```python
%timeit sum2(100)
```

 ## Big-O notation describes how quickly runtime will grow relative to the input as the input get arbitrarily large 

#### We want to compare how quickly runtime will grows not compare exact runtimes 
#### We want to compare for a variety of input sizes, we are opny concerned with runtime grow relative to the input.This is why we use n for notation











```python
import math 
```

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib in line
plt.style.use('bmh')
```

```python
n=np.linspace(1,10,1000)
labels=['Constant','Logarithmic','Linear','LogLinear','Quadratic','Cubic','Exponential']
big_o=[np.ones(n.shape),np.log(n),n,n*np.log(n),n**2,n**3,2**n]

plt.figure(figsize=(10,12))
plt.ylim(0.50)

for i in range(len(big_o)):
    plt.plot(n,big_o[i],label=labels[i]) 
    

plt.legend(loc=0)
plt.ylabel("Relative runtime")
plt.xlabel('n')
```
